```python
l = []
for i in range(15):
    l.append('__')
def printlist(list):
    for index, value in enumerate(l, start = 1):
        print(value, end = ('  '))
        if index %5 == 0:
            print("")
    print("")
    print("")

l1 = [1, 2, 3, 4, 5, 6, 7, 8]
n = len(l1)
r = 5
answer = []

def dfs(idx, list):
    if len(list) == r:
        answer.append(list[:])
        return
    
    for i in range(idx, n):
        dfs(i+1, list+[l1[i]])

dfs(0, [])

for i in answer:
    l = []
    for j in range(15):
        l.append('__')
    k = 0
    while k < len(i):
        l[(i[k]-1)*2] = '00'
        k += 1
    printlist(l)
```
